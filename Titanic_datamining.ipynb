```python
import pandas as pd
import math

csv_file = '/content/titanic.csv'

# Sample dataset (replace this with your dataset)
#data = pd.DataFrame({
   # 'Survived': [1, 0, 1, 0, 1, 0, 1, 1, 0, 1],
    #'Age': [22, 38, 26, 35, 35, 54, 2, 27, 14, 4],
    #'Pclass': [3, 1, 3, 1, 3, 2, 1, 3, 3, 2]
#})

data = pd.read_csv(csv_file)

# Function to calculate entropy
def calculate_entropy(data, target_attribute):
    total_samples = len(data)
    if total_samples == 0:
        return 0

    entropy = 0
    unique_labels = data[target_attribute].unique()

    for label in unique_labels:
        p_label = len(data[data[target_attribute] == label]) / total_samples
        entropy -= p_label * math.log(p_label, 2)

    return entropy

# Function to calculate information gain
def calculate_information_gain(data, attribute, target_attribute):
    total_samples = len(data)
    subset_entropy = 0.0

    unique_values = data[attribute].unique()
    for value in unique_values:
        subset = data[data[attribute] == value]
        subset_weight = len(subset) / total_samples
        subset_entropy += subset_weight * calculate_entropy(subset, target_attribute)

    information_gain = calculate_entropy(data, target_attribute) - subset_entropy
    return information_gain

# Calculate entropy for the "Survived" attribute
entropy_survived = calculate_entropy(data, 'Survived')

# Calculate information gain for other attributes
attributes = data.columns.tolist()
attributes.remove('Survived')  # Exclude "Survived" attribute
information_gains = {}

for attribute in attributes:
    information_gains[attribute] = calculate_information_gain(data, attribute, 'Survived')

# Print the results
print(f"Entropy for 'Survived': {entropy_survived:.2f}")
print("Information Gains:")
for attribute, gain in information_gains.items():
    print(f"{attribute}: {gain:.2f}")
```

```python
import pandas as pd
import numpy as np

class Node:
    def __init__(self, label=None, attribute=None, branches=None):
        self.label = label  # Class label if leaf node, None otherwise
        self.attribute = attribute  # Attribute to split on
        self.branches = branches if branches else {}  # Child nodes

def calculate_entropy(data):
    n = len(data)
    if n == 0:
        return 0
    p_1 = len(data[data['Survived'] == 1]) / n
    p_0 = len(data[data['Survived'] == 0]) / n
    if p_1 == 0 or p_0 == 0:
        return 0
    entropy = -p_1 * np.log2(p_1) - p_0 * np.log2(p_0)
    return entropy

def information_gain(data, attribute):
    total_entropy = calculate_entropy(data)
    unique_values = data[attribute].unique()
    weighted_entropy = 0

    for value in unique_values:
        subset = data[data[attribute] == value]
        weighted_entropy += (len(subset) / len(data)) * calculate_entropy(subset)

    return total_entropy - weighted_entropy

def id3(data, attributes, target_attribute):
    if len(data) == 0:
        return Node(label='No examples', attribute=None)
    elif len(data['Survived'].unique()) == 1:
        label = data['Survived'].iloc[0]
        return Node(label=label, attribute=None)
    elif len(attributes) == 0:
        label = data['Survived'].mode().iloc[0]
        return Node(label=label, attribute=None)
    else:
        best_attribute = max(attributes, key=lambda a: information_gain(data, a))
        root = Node(attribute=best_attribute)

        for value in data[best_attribute].unique():
            subset = data[data[best_attribute] == value]
            if len(subset) == 0:
                label = data['Survived'].mode().iloc[0]
                root.branches[value] = Node(label=label, attribute=None)
            else:
                new_attributes = attributes.copy()
                new_attributes.remove(best_attribute)
                root.branches[value] = id3(subset, new_attributes, target_attribute)

        return root

# Sample dataset (replace this with your dataset)
#data = pd.DataFrame({
    #'Survived': [1, 0, 1, 0, 1, 0, 1, 1, 0, 1],
    #'Pclass': [3, 1, 3, 1, 3, 2, 1, 3, 3, 2],
    #'Sex': ['male', 'female', 'female', 'female', 'male', 'male', 'female', 'female', 'male', 'female']
#})

csv_file = '/content/titanic.csv'
data = pd.read_csv(csv_file)

# Get the list of attributes (exclude 'Survived')
attributes = list(data.columns)
attributes.remove('Survived')

# Build the decision tree
tree = id3(data, attributes, 'Survived')

# Print the decision tree (in a simple textual format)
def print_tree(node, depth=0):
    if node.label is not None:
        print('  ' * depth + f'Class: {node.label}')
    else:
        print('  ' * depth + f'Attribute: {node.attribute}')
        for value, branch in node.branches.items():
            print('  ' * depth + f'Value: {value}')
            print_tree(branch, depth + 1)

print_tree(tree)
```

```python
import pandas as pd
import numpy as np
import math

class Node:
    def __init__(self, label=None, attribute=None, branches=None):
        self.label = label
        self.attribute = attribute
        self.branches = branches if branches else {}

def calculate_gini_index(data):
    total_samples = len(data)
    if total_samples == 0:
        return 0

    gini = 1
    unique_labels = data['Survived'].unique()

    for label in unique_labels:
        p_label = len(data[data['Survived'] == label]) / total_samples
        gini -= p_label ** 2

    return gini

def calculate_information_gain(data, attribute):
    total_samples = len(data)
    subset_gini = 0.0

    unique_values = data[attribute].unique()
    for value in unique_values:
        subset = data[data[attribute] == value]
        subset_gini += (len(subset) / total_samples) * calculate_gini_index(subset)

    information_gain = calculate_gini_index(data) - subset_gini
    return information_gain

def id3(data, attributes, target_attribute, max_depth):
    if len(data) == 0:
        return Node(label='No examples', attribute=None)
    elif len(data['Survived'].unique()) == 1:
        label = data['Survived'].iloc[0]
        return Node(label=label, attribute=None)
    elif len(attributes) == 0 or max_depth == 0:
        label = data['Survived'].mode().iloc[0]
        return Node(label=label, attribute=None)
    else:
        best_attribute = max(attributes, key=lambda a: calculate_information_gain(data, a))
        root = Node(attribute=best_attribute)
        max_depth -= 1

        for value in data[best_attribute].unique():
            subset = data[data[best_attribute] == value]
            if len(subset) == 0:
                label = data['Survived'].mode().iloc[0]
                root.branches[value] = Node(label=label, attribute=None)
            else:
                new_attributes = attributes.copy()
                new_attributes.remove(best_attribute)
                root.branches[value] = id3(subset, new_attributes, target_attribute, max_depth)

        return root

def predict_example(tree, example):
    if tree.label is not None:
        return tree.label
    else:
        attribute = tree.attribute
        value = example[attribute]
        if value in tree.branches:
            return predict_example(tree.branches[value], example)
        else:
            return data['Survived'].mode().iloc[0]

# Load your dataset from the CSV file
data = pd.read_csv('/content/titanic.csv')

# Determine the index to split the data
split_index = int(0.9 * len(data))

# Split the data into training and testing sets
train_data = data[:split_index]
test_data = data[split_index:]

# Get the list of attributes (exclude 'Survived')
attributes = list(data.columns)
attributes.remove('Survived')

# Build the decision tree
max_depth = 5  # Set the maximum depth of the tree
tree = id3(train_data, attributes, 'Survived', max_depth)

# Evaluate the accuracy of the decision tree on the testing data
correct_predictions = 0

for index, row in test_data.iterrows():
    prediction = predict_example(tree, row)
    if prediction == row['Survived']:
        correct_predictions += 1

accuracy = (correct_predictions / len(test_data)) * 100

# Calculate the Gini Index for the entire dataset
gini_index = calculate_gini_index(data)

# Calculate the information gain using Gini Index for all attributes
information_gains = {}
for attribute in attributes:
    information_gains[attribute] = calculate_information_gain(data, attribute)

# Print the results
print(f"Decision Tree Accuracy: {accuracy:.2f}%")
print(f"Gini Index for Entire Dataset: {gini_index:.2f}")
print("Information Gains:")
for attribute, gain in information_gains.items():
    print(f"{attribute}: {gain:.2f}")
```
